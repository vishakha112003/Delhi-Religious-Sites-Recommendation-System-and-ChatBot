## Creating column based on location

```python
keyword_mappings = {
    ('Chandni Chowk', 'Lal Qila', 'Jama Masjid','Daryaganj','Old Delhi'): 'Old Delhi',
    ('Saket','Qutb Minar','Chhatarpur','Hauz Khas','Outer Ring Road','Kailash',"Humayun's Tomb",'Lodhi','Defence','R. K. Puram','Chittaranjan Park','Begumpur','Khirki','South','Mehrauli','Kalkaji','Tughlakabad','Nizamuddin','Chirag Dilli',): 'South Delhi',
    ('Connaught Place','Mandi House','Pandit Pant Marg',"Lutyens'",'Delhi Race','Jhandewalan'): 'Central Delhi',
    ('Mayur Vihar','Pandav Nagar','Dilshad'):'East Delhi',
    ('Kashmere Gate','Gujranwala','Nangli Poona'):'North Delhi'
}

# Function to find region based on multiple keywords
def find_region(location):
    for keywords, region in keyword_mappings.items():
        for keyword in keywords:
            if keyword in location:
                return region
            
# Apply the function to create the 'Region' column
df['Region'] = df['Location'].apply(find_region)
```
